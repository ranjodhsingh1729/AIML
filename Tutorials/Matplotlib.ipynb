```python
t = np.arange(0, 5, 0.1)
y = [i for i in t]
y2 = [i**2 for i in t]
y3 = [i**3 for i in t]

print()
plt.plot(y, label="linear")
plt.plot(y2, label="square")
plt.plot(y3, label="cubic")

plt.grid(True)
plt.xlim(0, 40)
plt.ylim(0, 60)

plt.title("Polynomial Graphs")
plt.xlabel("Numbers")
plt.ylabel("Functions of Numbers")
plt.legend()
plt.show()
```

```python
t = np.arange(0, 5, 0.1)
y = [i for i in t]
y2 = [i**2 for i in t]
y3 = [i**3 for i in t]

print()
plt.plot(t, y, label="linear")
plt.plot(t, y2, label="square")
plt.plot(t, y3, label="cubic")

for i in range(0, 6):
    plt.text(i, i, (i, i), fontsize=10)
    plt.text(i, i**2, (i, i**2), fontsize=10)
    plt.text(i, i**3, (i, i**3), fontsize=10)

plt.grid(True)
plt.xlim(0, 5)
plt.ylim(0, 125)
plt.title("Polynomial Graphs")
plt.xlabel("Numbers")
plt.ylabel("Functions of Numbers")
plt.legend()

plt.show()
```
