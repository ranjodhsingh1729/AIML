```python
x = df[["age", "skullw"]].values
y = df["sex"].values
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(x[y==0, 0], y = x[y==0, 1], c='b', marker='s', label=0)
plt.scatter(x[y==1, 0], y = x[y==1, 1], c='r', marker='o', label=1)

plt.title("Possums")
plt.xlabel("age", fontsize=12)
plt.ylabel("skullw", fontsize=12)
plt.grid()
plt.legend(loc="upper right")
plt.show()
```
